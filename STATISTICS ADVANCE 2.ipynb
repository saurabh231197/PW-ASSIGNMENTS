Q1. What is hypothesis testing in statistics?
- Hypothesis testing is a statistical method used to make decisions about the properties of a population based on a sample. It involves formulating a null hypothesis (H0) and an alternative hypothesis (H1), then using sample data to determine whether to reject H0.
- **Example:** Testing whether a new drug is more effective than the current standard treatment.
-----------------------------------------------------------------
Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
- The null hypothesis (H0) is a statement that there is no effect or no difference, and it serves as the default assumption. The alternative hypothesis (H1) is a statement that there is an effect or a difference. Hypothesis testing aims to determine whether there is enough evidence to reject H0 in favor of H1.
- **Example:** H0: The new drug has no effect. H1: The new drug has an effect.
-----------------------------------------------------------------
Q3. What is the significance level in hypothesis testing, and why is it important?
- The significance level (alpha) is the probability of rejecting the null hypothesis when it is true. It is a threshold set by the researcher, commonly 0.05 or 0.01, to determine the strength of the evidence against H0. It is important because it controls the likelihood of making a Type I error.
- **Example:** Setting alpha to 0.05 means there is a 5% chance of rejecting H0 when it is true.
-----------------------------------------------------------------

Q4. What are Type 1 and Type 2 errors in hypothesis testing?
- A Type 1 error occurs when the null hypothesis is rejected when it is true (false positive). A Type 2 error occurs when the null hypothesis is not rejected when it is false (false negative). The significance level (alpha) controls the probability of a Type 1 error, while the power of the test affects the probability of a Type 2 error.
- **Example:** Type 1 error: Concluding a new drug is effective when it is not. Type 2 error: Concluding a new drug is not effective when it is.
-----------------------------------------------------------------
Q5. What is the difference between a Z-test and a T-test?Explain when to use
each.
- The Z-test is used when the population variance is known and the sample size is large, while the T-test is used when the population variance is unknown and the sample size is small. The T-test uses the T-distribution, which accounts for the additional uncertainty in estimating the population variance.
- **Example:** Z-test: Testing the mean of a large sample with known population variance. T-test: Testing the mean of a small sample with unknown population variance.

```python
"""Q6-: Write a Python program to generate a binomial distribution with n=10 and
p=0.5, then plot its histogram.
(Include your Python code and output in the code box below.)
Hint: Generate random number using random function."""

#ANSWER
import numpy as np
import matplotlib.pyplot as plt

def generate_and_plot_binomial():
    """
    Generates a binomial distribution and plots a histogram.

    n: The number of trials.
    p: The probability of success in a single trial.
    size: The number of experiments to simulate.
    """
    print("Generating and plotting a binomial distribution...")

    # Parameters for the binomial distribution
    n = 10  # Number of trials
    p = 0.5 # Probability of success
    size = 1000 # Number of experiments

    # Generate the binomial distribution data
    # The result is an array of random numbers, where each number represents the
    # number of successes in 'n' trials.
    binomial_data = np.random.binomial(n, p, size)

    # Create the histogram
    plt.figure(figsize=(10, 6))
    plt.hist(binomial_data, bins=np.arange(-0.5, n + 1.5, 1), edgecolor='black', alpha=0.7)

    # Set the labels and title for the plot
    plt.title(f'Binomial Distribution (n={n}, p={p})')
    plt.xlabel('Number of Successes')
    plt.ylabel('Frequency')

    # Ensure the x-axis ticks are integers corresponding to the number of successes
    plt.xticks(np.arange(0, n + 1, 1))

    # Add a grid for better readability
    plt.grid(axis='y', alpha=0.75)

    # Save the plot to a file
    plot_filename = 'binomial_distribution_histogram.png'
    plt.savefig(plot_filename)
    print(f"Plot saved successfully as '{plot_filename}'")
    print("A histogram showing the distribution of successes has been created.")

    # plt.show() # Uncomment this line to display the plot window directly

if __name__ == "__main__":
    generate_and_plot_binomial()
```

```python
""" Q7- Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9] """

#ANSWER
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Parameters
mu = 50  # population mean
sigma = 0.5  # population standard deviation
n = len(sample_data)
sample_mean = np.mean(sample_data)

# Z-statistic
z = (sample_mean - mu) / (sigma / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z)))

# Output
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Z-statistic: {z:.4f}")
print(f"P-value: {p_value:.4f}")

# Decision
if p_value < 0.05:
    print("Reject the null hypothesis: Sample mean is significantly different from population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference from population mean.")
```

```python
""" Q8- Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib. """

#ANSWER

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate data from a normal distribution
np.random.seed(42)  # for reproducibility
mu = 100            # true mean
sigma = 15          # true standard deviation
n = 200             # sample size

data = np.random.normal(mu, sigma, n)

# Step 2: Calculate sample mean and standard error
sample_mean = np.mean(data)
standard_error = stats.sem(data)  # standard error of the mean

# Step 3: Compute 95% confidence interval
confidence = 0.95
ci = stats.norm.interval(confidence, loc=sample_mean, scale=standard_error)

# Step 4: Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci[0]:.2f}, {ci[1]:.2f})")

# Step 5: Plot the data
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label='Sample Mean')
plt.axvline(ci[0], color='green', linestyle='dashed', linewidth=2, label='95% CI Lower')
plt.axvline(ci[1], color='green', linestyle='dashed', linewidth=2, label='95% CI Upper')
plt.title('Simulated Normal Distribution with 95% Confidence Interval')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
""" Q9-: Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean. """

#ANSWER
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    # Step 1: Calculate mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    # Step 2: Compute Z-scores
    z_scores = [(x - mean) / std_dev for x in data]

    # Step 3: Print summary
    print(f"Mean: {mean:.2f}")
    print(f"Standard Deviation: {std_dev:.2f}")
    print(f"First 5 Z-scores: {z_scores[:5]}")

    # Step 4: Plot histogram of Z-scores
    plt.figure(figsize=(10, 6))
    plt.hist(z_scores, bins=20, color='purple', edgecolor='black', alpha=0.7)
    plt.title('Histogram of Z-Scores')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label='Mean (Z=0)')
    plt.legend()
    plt.tight_layout()
    plt.show()

# Example usage
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

calculate_and_plot_z_scores(sample_data)
```
