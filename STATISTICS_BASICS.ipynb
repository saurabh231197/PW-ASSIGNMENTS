Q.1 What is the difference between descriptive statistics and inferential
statistics? Explain with examples.
Ans. Descriptive statistics and inferential statistics are two essential branches of statistical analysis, each serving a unique purpose. Descriptive statistics are used to summarize and present data in a meaningful way. They help in organizing raw data into understandable formats using measures such as mean, median, mode, range, and standard deviation. For example, if a teacher calculates the average marks of students in a class and displays the results using a bar graph, she is applying descriptive statistics to describe the performance of that specific group.
Inferential statistics, on the other hand, are used to make predictions or generalizations about a larger population based on a sample of data. This branch involves techniques such as hypothesis testing, confidence intervals, and regression analysis. For instance, if a researcher surveys a sample of 500 people to estimate the average income of an entire city, they are using inferential statistics to draw conclusions about the broader population. In summary, descriptive statistics help us understand and visualize the data we have, while inferential statistics allow us to make informed decisions and predictions beyond the data at hand.

Q.2 What is sampling in statistics? Explain the differences between random and stratified sampling.
Ans. Sampling in statistics is the process of selecting a subset of individuals or observations from a larger population to analyze and draw conclusions about the entire group. Since studying an entire population is often impractical due to time, cost, or logistical constraints, sampling allows researchers to gather meaningful insights efficiently and accurately.
There are various sampling methods, but two commonly used techniques are random sampling and stratified sampling. In random sampling, every member of the population has an equal chance of being selected. This method is simple and unbiased, making it ideal when the population is homogeneous. For example, if a researcher wants to survey 100 students from a university where all students have similar backgrounds, they might randomly select names from the student registry.
Stratified sampling, on the other hand, involves dividing the population into distinct subgroups or "strata" based on specific characteristics such as age, gender, income level, or education. Then, a random sample is taken from each stratum proportionally. This method ensures that all relevant subgroups are adequately represented in the final sample. For instance, if a researcher is studying consumer preferences across different age groups, they might divide the population into age brackets and randomly select individuals from each bracket to ensure balanced representation.
Q.3 : Define mean, median, and mode. Explain why these measures of central tendency are important.
Ans. Mean, median, and mode are fundamental measures of central tendency in statistics, used to describe the center or typical value within a dataset. The mean, often called the average, is calculated by summing all the values in a dataset and dividing by the number of values. It provides a general idea of the overall level of the data but can be influenced by extreme values or outliers. The median is the middle value when the data is arranged in ascending or descending order. If the dataset has an even number of values, the median is the average of the two middle numbers. It is especially useful when the data is skewed, as it is not affected by outliers. The mode is the value that appears most frequently in the dataset. It is helpful for identifying the most common or popular item, especially in categorical data.
These measures are important because they help simplify complex data sets by identifying a single representative value. They allow researchers, analysts, and decision-makers to quickly understand the general trend or behavior of the data. For example, in education, the mean score of a class can indicate overall performance, while the median can show the central achievement level, and the mode can highlight the most common score. Together, these measures provide a comprehensive view of the data’s distribution and are essential tools in fields ranging from economics and psychology to business and public health.

Q.4 Explain skewness and kurtosis. What does a positive skew imply about the data?
Ans. Skewness and kurtosis are statistical measures that describe the shape and characteristics of a data distribution beyond basic averages.
Skewness refers to the asymmetry of a distribution. If a dataset is perfectly symmetrical, it has zero skewness. A positive skew means the distribution has a longer tail on the right side, indicating that most values are concentrated on the lower end, with a few unusually high values pulling the mean to the right. For example, income data often shows positive skew because while most people earn moderate incomes, a few individuals earn extremely high amounts.
Kurtosis, on the other hand, measures the "tailedness" or the sharpness of the peak of a distribution. It tells us how heavily the tails of the distribution differ from a normal distribution. A high kurtosis (leptokurtic) indicates more data in the tails and a sharper peak, suggesting a higher likelihood of outliers. A low kurtosis (platykurtic) implies a flatter distribution with fewer extreme values.
In summary, a positive skew implies that the data has a concentration of lower values with a few high outliers, while kurtosis helps assess the presence and impact of those outliers on the overall distribution. These measures are crucial for understanding the behavior of data, especially when making decisions based on statistical models.

```python
# Q.7 Compute the covariance and correlation coefficient between the following
#two datasets provided as lists in Python:
#list_x = [10, 20, 30, 40, 50]
#list_y = [15, 25, 35, 45, 60]
#ANS.

import numpy as np

# Given datasets
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Convert lists to NumPy arrays
x = np.array(list_x)
y = np.array(list_y)

# Compute covariance matrix
cov_matrix = np.cov(x, y, ddof=0)
covariance = cov_matrix[0][1]

# Compute correlation coefficient
correlation = np.corrcoef(x, y)[0][1]

# Display results
print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

```python
#Q.8 Write a Python script to draw a boxplot for the following numeric list and
#identify its outliers. Explain the result:
#data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]
#ANS.

import matplotlib.pyplot as plt
import numpy as np

# Data list
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Create boxplot
plt.boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title('Boxplot of Numeric Data')
plt.xlabel('Value')
plt.show()

# Identify outliers using IQR
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("Outliers:", outliers)
```

```python
# Q.8 You are working as a data analyst in an e-commerce company. The
#marketing team wants to know if there is a relationship between advertising spend and
#daily sales.
#● Explain how you would use covariance and correlation to explore this
#relationship.
#● Write Python code to compute the correlation between the two lists:
#advertising_spend = [200, 250, 300, 400, 500]
#daily_sales = [2200, 2450, 2750, 3200, 4000]

#ANS.
"""Covariance measures how two variables change together. A positive covariance
indicates that as advertising spend increases, daily sales tend to increase too
However, it doesn't tell us the strength or consistency of the relationship.

Correlation coefficient (typically Pearson’s r) standardizes this relationship on a
scale from -1 to +1. A value close to +1 means a strong positive linear relationship, 0
means no linear relationship, and -1 means a strong negative linear relationship.

By calculating both, we can determine not just whether there's a relationship, but how strong and consistent it is """

import numpy as np

# Data lists
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Convert to NumPy arrays
x = np.array(advertising_spend)
y = np.array(daily_sales)

# Compute covariance
cov_matrix = np.cov(x, y, ddof=0)
covariance = cov_matrix[0][1]

# Compute correlation coefficient
correlation = np.corrcoef(x, y)[0][1]

# Display results
print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)

```

```python
#Q.9 : Your team has collected customer satisfaction survey data on a scale of
#1-10 and wants to understand its distribution before launching a new product.
#● Explain which summary statistics and visualizations (e.g. mean, standard
#deviation, histogram) you’d use.
#● Write Python code to create a histogram using Matplotlib for the survey data:
#survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

#ANS.

""" To analyze the distribution of customer satisfaction scores (rated on a scale of 1–10), we use both summary statistics and visualizations
Mean: Measures the average satisfaction score.
Median: Indicates the middle score when data is sorted.
Mode: Shows the most frequently occurring score.
Standard Deviation: Reflects how spread out the scores are.
Minimum and Maximum: Help identify the range of responses.
Count: Total number of responses collected.

A histogram is ideal for visualizing the frequency distribution of scores.
It shows how many customers gave each score or range of scores, helping us spot trends like clustering or skewness.
This helps visualize how customer satisfaction scores are distributed,
making it easier to identify patterns such as high satisfaction clusters or areas needing improvement."""

import matplotlib.pyplot as plt

# Survey data
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Create histogram
plt.hist(survey_scores, bins=7, edgecolor='black', color='skyblue')
plt.title('Customer Satisfaction Score Distribution')
plt.xlabel('Satisfaction Score')
plt.ylabel('Number of Responses')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.show()




```
