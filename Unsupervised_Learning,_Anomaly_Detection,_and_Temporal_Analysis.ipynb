Q1. What is Dimensionality Reduction? Why is it important in machine
learning?
ANS. Dimensionality reduction involves transforming data from a high-dimensional space into a lower-dimensional one. This is done by eliminating redundant or less informative features while retaining the core structure and patterns in the data. Common techniques include:
- Principal Component Analysis (PCA): Projects data onto a smaller number of orthogonal axes that capture the most variance.
- Linear Discriminant Analysis (LDA): Focuses on maximizing class separability.
- t-SNE and UMAP: Non-linear techniques for visualizing high-dimensional data in 2D or 3D.

Dimensionality reduction is a key step in many ML workflows as:
- Reduces Overfitting: Fewer features mean less chance for the model to learn noise instead of signal.
- Improves Model Performance: Less complex models often train faster and generalize better.
- Enhances Visualization: Helps in visualizing high-dimensional data in 2D or 3D for exploratory analysis.
- Lowers Computational Cost: Reducing dimensions cuts down on memory usage and speeds up training and inference.
- Removes Multicollinearity: Helps eliminate correlated features that can distort model predictions.
For example, in a dataset predicting house prices, including too many features like “type of doorknob” or “color of curtains” might not add value and could confuse the model. Dimensionality reduction helps focus on impactful features like square footage, location, and number of bedrooms


Q2. Name and briefly describe three common dimensionality reduction
techniques..

ANS.
three widely used dimensionality reduction techniques in machine learning:
1.  Principal Component Analysis (PCA)
- Purpose: Transforms the original features into a new set of uncorrelated variables called principal components.
- How it works: It identifies directions (components) that capture the maximum variance in the data.
- Use case: Ideal for continuous data and preserving global structure.
2. t-Distributed Stochastic Neighbor Embedding (t-SNE)
- Purpose: Visualizes high-dimensional data by mapping it into 2D or 3D space.
- How it works: Focuses on preserving local relationships between data points using probabilistic similarity.
- Use case: Great for visualizing clusters in complex datasets like images or text embeddings.
3. Linear Discriminant Analysis (LDA)
- Purpose: Reduces dimensions while maximizing class separability.
- How it works: Projects data onto a lower-dimensional space that best separates different classes.
- Use case: Common in supervised learning tasks like classification.

clustering algorithms.

ANS. Clustering in unsupervised learning is the process of grouping similar data points together based on patterns or similarities without using labeled data. It helps uncover hidden structures in datasets and is widely used in data exploration, customer segmentation, image analysis, and anomaly detection.

Three widely used clustering methods are:

1-K-Means - Partitions data into k clusters by minimizing intra-cluster variance
used for: Large datasets with spherical clusters

2-Hierarchical Clustering - Builds a tree (dendrogram) of clusters using either agglomerative or divisive methods.
used for: Visualizing nested clusters

3-DBSCAN - Groups points based on density; identifies noise and outliers.
used for: Arbitrary-shaped clusters and noisy data

Q4.  Explain the concept of anomaly detection and its significance.

ANS. Anomaly detection (also known as outlier detection) involves recognizing instances in data that do not conform to expected patterns. These anomalies can be:
- Point anomalies: A single data point significantly differs from the rest (e.g., a sudden spike in temperature).
- Contextual anomalies: A data point is anomalous in a specific context (e.g., high temperature in winter).
- Collective anomalies: A group of related data points is anomalous when considered together (e.g., a sequence of failed login attempts indicating a cyberattack).

Anomaly detection is vital across many domains due to its ability to:
- Detect fraud: Identify unusual transactions in banking or e-commerce that may indicate fraud.
- Monitor system health: Spot hardware failures or software bugs in IT systems and industrial equipment.
- Enhance cybersecurity: Detect unauthorized access, malware, or network intrusions.
- Ensure quality control: Identify defects in manufacturing processes.
- Improve business operations: Reveal inefficiencies or unexpected trends in customer behavior or supply chains.




Q5. List and briefly describe three types of anomaly detection techniques

ANS. Three common types of anomaly detection techniques used in machine learning:

1. Statistical Methods
- How it works: Assumes that normal data follows a known distribution (e.g., Gaussian). Points that fall far from the mean (e.g., beyond 3 standard deviations) are flagged as anomalies.
- Best for: Simple datasets with well-understood distributions.
- Example: Z-score, Grubbs’ test.
2. Machine Learning-Based Methods
- How it works: Uses models to learn patterns in data and identify deviations. Can be supervised (e.g., classification) or unsupervised (e.g., clustering).
- Best for: Complex, high-dimensional data where statistical assumptions may not hold.
- Example: Isolation Forest, One-Class SVM, Autoencoders.


3. Proximity-Based Methods
- How it works: Measures the distance or density of data points. Anomalies are those far from others or in low-density regions.
- Best for: Data where similarity or distance metrics are meaningful.
- Example: k-Nearest Neighbors (k-NN), DBSCAN.



Q6.: What is time series analysis? Mention two key components of time series
data

ANS. Time series data consists of observations recorded sequentially over time—daily stock prices, monthly sales figures, or hourly temperature readings, for example. Time series analysis involves:
- Understanding the underlying structure and function of the data.
- Modeling patterns such as trends and seasonality.
- Forecasting future values based on historical behavior.
This analysis helps in making informed predictions and strategic decisions.

Two key components are
 1- Trend
- Represents the long-term progression or direction in the data.
- It shows whether the data is increasing, decreasing, or remaining stable over time.
- Example: A steady rise in housing prices over several years.
 2- Seasonality
- Refers to periodic fluctuations that occur at regular intervals due to seasonal factors.
- These patterns repeat over a known, fixed period such as daily, monthly, or yearly.
- Example: Increased retail sales during the holiday season.


Q7. Describe the difference between seasonality and cyclic behavior in time
series.

ANS. Seasonality and cyclic behavior are both patterns observed in time series data, but they differ in their frequency, predictability, and underlying cause

Seasonality
- Definition: Regular, repeating patterns that occur at fixed intervals due to seasonal factors.
- Frequency: Occurs at known and consistent periods (e.g., daily, weekly, monthly, yearly).
- Cause: Driven by calendar-related influences like weather, holidays, or business cycles.
- Example: Retail sales peaking every December due to holiday shopping.

Cyclic Behavior
- Definition: Fluctuations that occur over longer, variable periods, often tied to economic or natural cycles.
- Frequency: Irregular and not fixed; cycles can span months to years.
- Cause: Influenced by broader factors like economic expansions and recessions, or commodity price cycles.
- Example: Housing market booms and busts over a decade.



```python
""" Q8.Write Python code to perform K-means clustering on a sample dataset.
(Include your Python code and output"""
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# --- Configuration ---
N_SAMPLES = 1000  # Total number of data points
N_FEATURES = 2    # Dimension of the data points (for 2D plotting)
N_CLUSTERS = 4    # The number of clusters we want to find
RANDOM_STATE = 42 # Seed for reproducibility

# --- 1. Data Generation ---
# We use make_blobs to create a synthetic dataset that clearly exhibits cluster structure.
print(f"Generating synthetic data with {N_SAMPLES} samples and {N_CLUSTERS} true clusters...")
X, y_true = make_blobs(
    n_samples=N_SAMPLES,
    centers=N_CLUSTERS,
    n_features=N_FEATURES,
    cluster_std=1.0, # Standard deviation of the clusters
    random_state=RANDOM_STATE
)

# X contains the features (coordinates of the points)
# y_true contains the true, underlying cluster labels (which we ignore during clustering)

# --- 2. K-Means Clustering ---
# Initialize the KMeans model
# n_init='auto' ensures we run the algorithm multiple times with different centroid initializations
# to find the best result.
kmeans = KMeans(
    n_clusters=N_CLUSTERS,
    random_state=RANDOM_STATE,
    n_init='auto'
)

# Fit the model to the data (this is where the clustering happens)
print("Fitting K-Means model...")
kmeans.fit(X)

# Get the computed cluster labels for each data point
y_kmeans = kmeans.predict(X)

# Get the coordinates of the final cluster centers
centers = kmeans.cluster_centers_

print(f"Clustering complete. Found {len(centers)} cluster centers.")
print("Cluster Centers (Coordinates):")
for i, center in enumerate(centers):
    print(f"  Cluster {i}: ({center[0]:.2f}, {center[1]:.2f})")

# --- 3. Visualization ---
plt.figure(figsize=(10, 7))

# Plot the data points, colored according to the cluster predicted by K-Means
# We loop through the predicted clusters (0 to N_CLUSTERS-1)
for i in range(N_CLUSTERS):
    # Select points belonging to the current cluster 'i'
    cluster_points = X[y_kmeans == i]

    plt.scatter(
        cluster_points[:, 0], # X-coordinates
        cluster_points[:, 1], # Y-coordinates
        s=50,                 # Marker size
        label=f'Cluster {i}', # Legend label
        alpha=0.6             # Transparency
    )

# Plot the cluster centers (the centroids)
plt.scatter(
    centers[:, 0],
    centers[:, 1],
    marker='X', # Use 'X' marker for centers
    s=200,      # Larger size
    c='red',    # Red color
    label='Centroids',
    edgecolor='black',
    linewidth=1.5
)

plt.title(f'K-Means Clustering Results (K={N_CLUSTERS})')
plt.xlabel('Feature 1 (X)')
plt.ylabel('Feature 2 (Y)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(loc='lower left')
plt.show()

print("\nVisualization displayed. The 'X' markers indicate the final cluster centers.")

# Optional: Show how K-Means performance compares to the true labels (if available)
print("\n--- Model Evaluation (compared to true labels) ---")
from sklearn.metrics import adjusted_rand_score
ari = adjusted_rand_score(y_true, y_kmeans)
print(f"Adjusted Rand Index (ARI): {ari:.4f} (Closer to 1.0 is better)")
```

```python
""" Q9. What is inheritance in OOP? Provide a simple example in Python."""
""" ANS. Inheritance is a fundamental concept in Object-Oriented Programming (OOP) that allows a new class (the child or derived class)
to inherit properties and methods from an existing class (the parent or base class)."""
# 1. Parent Class (Base Class)
class Animal:
    def __init__(self, name, species):
        # Initialize attributes common to all animals
        self.name = name
        self.species = species

    def make_sound(self):
        # A generic method that can be overridden
        print(f"{self.name} the {self.species} makes a sound.")

    def introduce(self):
        # Common behavior
        print(f"Hi, I'm {self.name}. I am a {self.species}.")

# 2. Child Class (Derived Class)
# The class name 'Dog' followed by '(Animal)' in the definition signifies inheritance.
class Dog(Animal):
    def __init__(self, name, breed):
        # Call the parent's constructor to initialize 'name' and 'species'
        # 'super()' refers to the parent class (Animal)
        super().__init__(name, species="Canine")
        self.breed = breed  # Add Dog's unique attribute

    # Override the parent's method for specific behavior
    def make_sound(self):
        print("Woof! Woof!")

    # Add Dog's unique method
    def fetch(self):
        print(f"{self.name} is fetching the ball.")

# --- Usage ---
my_dog = Dog("Buddy", "Golden Retriever")

# Inherited methods from Animal:
my_dog.introduce() # Output: Hi, I'm Buddy. I am a Canine.

# Overridden method:
my_dog.make_sound() # Output: Woof! Woof! (Dog's specific sound)

# Unique method of Dog:
my_dog.fetch() # Output: Buddy is fetching the ball.

# Demonstrating the generic Animal sound:
generic_animal = Animal("Leo", "Lion")
generic_animal.make_sound() # Output: Leo the Lion makes a sound.
```
